## Dicionário de Variáveis

*`total_cases` - Total de casos confirmados.
*`new_cases` - Novos casos confirmados.

*`new_cases_smoothed_per_million` - Novos casos por 1 milhão de pessoas nos últimos 7 dias.

*`new_deaths_per_million` - Novos mortes por 1 milhão de pessoas.

*`reproduction_rate` - Estimativa em tempo real da efetiva taxa de reprodução da COVID-19.

### EXTRA: Será que o PIB *per capita* de um país está relacionado com o seu número total de mortes por COVID-19?

Hmmm, interessante! Vemos pelo gráfico acima que **aparentemente** as variáveis mostradas não apresentam uma forte correlação. Se não sabe o que é correlação fique em paz que eu já escrevi um outro artigo contemplando esse tema! Basta procurar a seção *\"Qual a correlação existente entre as variáveis?"* [aqui](https://medium.com/@peubene/analisando-os-dados-do-airbnb-tokyo-5a73e825be17). 

Vamos investigar esse possível relacionamento com um pouco mais de detalhe, usando o método `corr()`, *capaz de calcular correlação entre pares de variáveis, excluindo valores ausentes/nulos*. AH! Quase esqueci de comentar, mas aqui testarei três métodos de cálculo de correlação, que são apresentados de forma introdutória [nesse excelente blog](https://datascience.stackexchange.com/questions/64260/pearson-vs-spearman-vs-kendall/64261). Esse é um assunto MUITO extenso e poderíamos ter um projeto inteiro dedicado à temática de correlação, mas deixaremos essas discussões para outros momentos.

>Provavelmente, você está pensando o mesmo que eu: a biblioteca Pandas é a melhor invenção na história da humanidade (depois do ar condicionado, é claro kkkk). Brincadeiras a parte, mãos à obra!



```
x = df_paises.gdp_per_capita
y = df_paises.total_deaths

# usando método de Pearson (parâmetro r)
print(f"Pearson - Parâmetro R:\t\t{x.corr(y, method='pearson')}")

# usando método de Spearmean (parâmetro pho)
print(f"Spearmean - Parâmetro Pho:\t{x.corr(y, method='spearman')}")

# usando método de Kendall's (parâmetro tau)
print(f"Kendall - Parâmetro Tau:\t{x.corr(y, method='kendall')}")
```

Todos os métodos testados apontam para a mesma conclusão: a correlação entre as variáveis é fraca, o que (apesar de não ser conclusivo) reforça a intuição de que  o PIB *per capita* não está fortemente relacionado com o número total de mortes no país.
> Caso você queira saber mais sobre as peculiaridades e diferenças associadas à cada um desses métodos, recomendo fortemente ler [esse artigo](https://realpython.com/numpy-scipy-pandas-correlation-python/#visualization-of-correlation) do blog **Real Python**.

#### **Criando matrizes de correlação**

Nesse caso, **temos muitas variáveis para serem analisadas**, o que pode vir a prejudicar a visualização na matriz. Por isso, vamos implementar uma forma de **extrair as correlações mais significativas** e suprimir as não tão relevantes.

Além disso, vimos anteriormente que existem diferentes métodos para o cálculo da correlação entre variáveis: `pearson`, `spearman` e `kendall`. **Mas em qual contexto devo usar cada um desses métodos?**

<center><img alt="Colaboratory logo" width="35%"" src="https://raw.githubusercontent.com/Pedro1-21GW/sigmoidal/main/resources_proj2/duvidas2.png"></center>

Para nossa análise, é recomendado escolher um método que possua **maior generalidade** do que `pearson`, afinal podemos estar lidando com correlações não-lineares entre nossas *features* e dados com distribuição *não-normal*. Diante dessas limitações, é mais sensato utilizar métodos *não paramêtricos*, ou seja: não assumem distribuições para o conjunto de dados utilizado (apesar de serem gerais, ainda possuem algumas limitações como a pressuposição de [monotonicidade](https://pt.wikipedia.org/wiki/Fun%C3%A7%C3%A3o_mon%C3%B3tona)). Com tudo isso em mente, entramos no escopo dos métodos de `spearman` e `kendall`, que são os dois coeficientes de correlação por rank mais utilizados. Escolhi utilizar o método de `kendall` nessa breve análise.


> Sei que essa parte esta um pouco mais densa (especialmente se você não possue base/formação em estatística), por isso **recomendo fortemente**:
* [assistir esse vídeo.](https://www.youtube.com/watch?v=6uu4sFl1avE)
* e uma leitura dinâmica da seção *background* [desse artigo.](https://bmcbioinformatics.biomedcentral.com/articles/10.1186/s12859-015-0697-7)


##### Variável Foco: `total_deaths`

```
# vamos criar uma matriz de correlação
correlation = df_brasil.corr(method = 'kendall')

# número de variáveis a serem selecionadas
num_variables = 9

# vamos encontrar as "k" variáveis com maior módulo de coeficiente de correlação
cols = correlation.abs().nlargest(num_variables, 'total_deaths')['total_deaths']
cols
```

Vale mencionar que: a selecão realizada acima contemplou as variáveis com os 9 maiores **valores absolutos** do coeficiente de Pearson, considerando assim a **relevância das correlações negativas** em nossa análise.

Além disso, como estamos trabalhando como muitas variáveis, observar uma matriz de correlação com dimensões 9 X 9 pode ser um pouco cansativo... por isso vou aproveitar a simetria que temos em relação à diagonal principal de nossa matriz, plotando uma **visualização mais enxuta e sem perdas**, logo abaixo. 

```
# extraindo o label indentificador das linhas apresentadas acima
index_cols = cols.index

# gerando uma nova matriz de correlação com as colunas selecionadas acima
correlation_selection = df_brasil[index_cols].corr(method = 'kendall')

# heatmap para a matriz criada 
fig, ax = plt.subplots(figsize = (10,10))
sns.set(font_scale = 1.25)

# vamos plotar apenas a metade inferior, por isso vamos antes criar uma máscara
mask = np.triu(np.ones_like(correlation_selection, dtype = bool))

sns.heatmap(correlation_selection, mask = mask, fmt = '.2f', annot = True, square = True, 
            vmax=1,cmap='RdBu')

ax.set_title("Correlation Matrix | Focus on Total Deaths")
plt.tight_layout()
```

Podemos comentar algumas correlações interessantes na matriz acima:
* Na parte superior, podemos ver um grande conjunto de variáveis com correlação positiva *praticamente* perfeita. Dessa forma, podemos concluir que nossa variável foco, `total_deaths`, *\"caminha junto"* com `total_tests | total_vaccinations | people_vaccinated | total_cases `, o que é equivalente a afirmar que se trata de um relacionamento monotonicamente crescente. 
> Obs: Note, pela primeira célula do tópico, que correlações como 0.999993... estão sendo representadas aqui como 1.00.
* Em seguida, podemos comentar sobre a significativa correlação negativa (-0.54) entre a variável `stringency_index` e `total_deaths`, na medida em que observou-se uma diminuição no número total de casos para um crescimento no Índice de Restrição de Resposta Governamental (parâmetro com 0 = resposta mais branda e 100 = resposta mais estrita), em outras palvaras: ao passo em que medidas restritivas entraram em ação, com regulações mais severas, observou-se uma diminuição correspondente na taxa de crescimento no número de mortes... **MAS CUIDADO!**. **Apesar de ser contra-intuitivo: isso não aponta uma relação clara de causa-efeito entre medidas restritivas mais severas e uma diminuição no número de mortes!**
>  Para te provar que correlações não implicam em causalidade, acessem [esse site](https://www.tylervigen.com/spurious-correlations), em que podemos ver algumas *correlações espúrias*, que basicamente apresentam altos valores de correlação para coisas totalmente dissociadas (algumas chegam a ser cômicas), como por exemplo:
